# 冒泡排序

```python
def bubblesort(nums:list):
    n = len(nums)

    for end in range(n-1,-1,-1):
        for i in range(end):
            if nums[i] >= nums[i+1]:
                # swap(i, i+1)
                print(f"swap {nums[i]} with {nums[i+1]}")
                nums[i], nums[i+1] = nums[i+1], nums[i]
                print(f"now the nums is:{nums}")
        print("inner loop is over.")
    print("outer loop is over")
```

```python
# 测试BubbleSort:
nums1 = [5,3,4,1,2]
bubblesort(nums1)
```

# 树的遍历访问

对于三种遍历序列，存在采用命令栈表示的的迭代算法，这种算法的优点是三种风格统一代码简洁，缺点是时间复杂度在常数意义下较大。

```python
def traverse(root, order):
    if root is None:
        return

    stack = [(root, False)]

    while stack:
        node, visited = stack.pop()

        if node is None:
            continue

        if visited:
            print(node.data, end=" ")
            continue

        if order == "pre":
            # 目标顺序：根、左、右
            stack.append((node.right, False))
            stack.append((node.left, False))
            stack.append((node, True))

        elif order == "in":
            # 目标顺序：左、根、右
            stack.append((node.right, False))
            stack.append((node, True))
            stack.append((node.left, False))

        elif order == "post":
            # 目标顺序：左、右、根
            stack.append((node, True))
            stack.append((node.right, False))
            stack.append((node.left, False))
```
